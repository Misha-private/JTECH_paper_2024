## Example of filter response for variaous schemes

```python
#
# plot_x.py
#

from google.colab import drive
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
drive.mount('/content/drive')

x = np.arange(0,792,1)
y1 = np.loadtxt('/content/drive/My Drive/MAP_prep/v1_x.dat')
y2 = np.loadtxt('/content/drive/My Drive/MAP_prep/v2_x.dat')
y3 = np.loadtxt('/content/drive/My Drive/MAP_prep/v3_x.dat')
y4 = np.loadtxt('/content/drive/My Drive/MAP_prep/v4_x.dat')
y5 = np.loadtxt('/content/drive/My Drive/MAP_prep/v5_x.dat')
y6 = np.loadtxt('/content/drive/My Drive/MAP_prep/v6_x.dat')

fig = plt.figure(figsize=(6,5))

ax = plt.axes()
ax.set_facecolor('antiquewhite')

plt.plot(x,y1, color='#DC3220', linestyle='--', label='$v_1$')
plt.plot(x,y2, color='#D41159', linestyle=':', label='$v_2$')
plt.plot(x,y3, color='#40B0A6', linewidth=2, linestyle='--', label='$v_3$')
plt.plot(x,y4, label='$v_4$')
plt.plot(x,y5, color='#E66100',label='$v_5$')
#plt.plot(x,y6, label='v6')

ax.legend(loc='upper right')

ax.grid(True)
ax.set_xlim([310,490])
plt.title('Line cross-section in x-direction', ha='center', fontsize=14)
plt.xlabel('Number of grid points in $x$-direction')
plt.ylabel('Filter response')
#plt.title('Radial filter horizontally, line filter vertically')
plt.savefig('x-cross_section.png')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([64, 88, 168, 256, 336, 448, 512, 704, 768,
924, 1056, 1408, 1848, 2464, 2816, 4224, 5632, 7392, 9856,11264])

y_rad= np.array([39.0414, 37.8965, 19.6915, 13.5923, 10.0216, 7.5942,
6.5301, 4.9975, 4.0391, 3.9416, 3.4705, 2.8476, 2.1093, 1.6565, 1.4811,
1.0542, 0.8534, 0.6821, 0.5646, 0.4533 ])

y_lin= np.array([38.1472, 28.4601, 15.5801, 10.2618, 8.039, 6.0147,
5.3309, 4.1332, 3.4184, 3.3174, 2.9878, 2.3972, 1.9156, 1.5684,
1.4369, 1.0654, 0.9113, 0.7444, 0.6398, 0.5429])

x_log = np.log10(x)

y_rad_inv = 1/y_rad
y_lin_inv = 1/y_lin
y_rad_log = np.log10(y_rad_inv)
y_lin_log = np.log10(y_lin_inv)

fig = plt.figure(figsize=(8, 4))

ax = plt.axes()
ax.set_facecolor('antiquewhite')

plt.plot(x_log, y_rad_log, 'bo', label='radial')
plt.plot(x_log,y_rad_log, 'b',  linewidth=0.5)

plt.plot(x_log, y_lin_log, 'ro', label='line')
plt.plot(x_log,y_lin_log, 'r',  linewidth=0.5)
plt.grid()
plt.title('Logarithm (base 10) of speedup of 2D+1D radial and line filters')
plt.xlabel('$log_{10}(Number \, of \, PEs)$')
plt.ylabel('$log_{10}(Speedup)$')
plt.legend(loc='best')

plt.savefig('2d1d.png')

plt.show()
```

```python
y_gain = (y_rad-y_lin)/y_rad*100
print(y_gain)
```

```python
fig = plt.figure(figsize=(8, 4))

ax = plt.axes()
ax.set_facecolor('aliceblue')

plt.plot(x_log, y_gain, 'o', color=(0.9, 0.2, 0.5, 0.5), label='Gain 2D+1D')
plt.plot(x_log,y_gain, 'b',  linewidth=0.5)


plt.grid()
plt.title('Gain (1 - Line/Radial)%')
plt.xlabel('$log_{10}(Number \, of \, PEs)$')
plt.ylabel('%')
plt.legend(loc='best')

plt.savefig('gain_2d1d.png')

plt.show()
```

## Scaling from GSI

```python
import numpy as np
import matplotlib.pyplot as plt


x_red = np.array([168, 256, 336, 448, 512, 704, 768, 924, 1056, 1408])
x_red_log = np.log10(x_red)
y_rf= np.array([0.005, 0.0105, 0.0145, 0.0167, 0.0171, 0.0177, 0.0180, 0.019, 0.0192, 0.0192 ])
y_mgbf = np.array([0.005, 0.0120, 0.0172, 0.0220, 0.0261, 0.0382, 0.0410, 0.044, 0.0491, 0.0540 ])
y_mgbf_log = np.log10(y_mgbf)
y_rf_log = np.log10(y_rf)


fig = plt.figure(figsize=(8, 4))

ax = plt.axes()
ax.set_facecolor('floralwhite')

plt.plot(x_red_log, y_mgbf_log, 'o', color =  '#E66100')
plt.plot(x_red_log, y_mgbf_log,  color =  'black', label = 'MGBF',linewidth=0.75)

plt.plot(x_red_log, y_rf_log, 'o', color =  '#40B0A6')
plt.plot(x_red_log, y_rf_log, color = '#DC3220', label='RF', linewidth=0.75)


plt.grid()
plt.title('Logarithm (base 10) of speedup in GSI of radial MGBF and RF schemes')

plt.xlabel('$log_{10}(Number \, of \, PEs)$')
plt.ylabel('$log_{10}(Speedup)$')

plt.legend(loc='best')
plt.savefig('log_speedup_mgf_rf.png')

plt.show()
```
